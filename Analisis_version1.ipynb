```python
columnas_validar = ['heart_rate', 'respiratory_rate', 'spo2']
dd[columnas_validar] = dd[columnas_validar].apply(pd.to_numeric, errors='coerce')

#Creamos las validaciones (True si es normal, False si es inválido/nulo)
# Frecuencia cardiaca: 60 - 100
dd['fc_valida'] = dd['heart_rate'].between(60, 100)

# Frecuencia respiratoria: 12 - 20
dd['fr_valida'] = dd['respiratory_rate'].between(12, 20)

# Saturación de oxígeno: 95 - 100
dd['spo2_valida'] = dd['spo2'].between(95, 100)

# 3. Contar cuántos datos inválidos hay por cada signo
print("Conteo de registros VÁLIDOS por categoría:")
print(dd[['fc_valida', 'fr_valida', 'spo2_valida']].sum())
```

```python
dd_limpio = dd[dd['fc_valida'] & dd['fr_valida'] & dd['spo2_valida']]

print(f"Registros totales originales: {len(dd)}")
print(f"Registros con datos fisiológicamente válidos: {len(dd_limpio)}")
```

```python
dd = dd.replace(r'^\s*$', np.nan, regex=True)
```
